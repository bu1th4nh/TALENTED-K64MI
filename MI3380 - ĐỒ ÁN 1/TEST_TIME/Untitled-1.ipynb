```python
import numpy as np
import pandas as pd 
import seaborn as sns

n_10 = pd.read_csv("log_10.csv");
n_50 = pd.read_csv("log_50.csv");
n_100 = pd.read_csv("log_100.csv");
n_200 = pd.read_csv("log_200.csv");
n_500 = pd.read_csv("log_500.csv");
n_1000 = pd.read_csv("log_1000.csv");
n_5000 = pd.read_csv("log_5000.csv");

master = [
    n_10,
    n_50,
    n_100,
    n_500,
    n_1000,
    n_5000
];

fig_siz = [12.8,7.2]
fig_dpi = 150
hs = 0.4
ws = 0.4
```

```python
for df in master:
    time1 = df.values[:,3];
    ndfs1 = df.values[:,4];
    notl1 = df.values[:,5];
    time2 = df.values[:,6];
    ndfs2 = df.values[:,7];
    notl2 = df.values[:,8];

    print(df.values[3,0], " & ", np.mean(time1), " & ", np.mean(ndfs1), " & ", np.mean(notl1), " & ", np.mean(time2), " & ", np.mean(ndfs2), " & ", np.mean(notl2), "\\\\ \hline");
    
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=fig_siz, dpi=fig_dpi)
gs  = fig.add_gridspec(2, 3, hspace=hs, wspace=ws)
axs = gs.subplots()
fig.suptitle("Thống kê thời gian chạy (giây) với thuật toán đường tăng trưởng")

idx = 3
for i in range(2):
    for j in range(3):
        axs[i, j].boxplot(master[i*3+j].values[:,idx], meanline=True)
        axs[i, j].set_title("n = {0}".format(master[i*3+j].values[3,0]))
```

```python
fig = plt.figure(figsize=fig_siz, dpi=fig_dpi)
gs  = fig.add_gridspec(2, 3, hspace=hs, wspace=ws)
axs = gs.subplots()
fig.suptitle("Thống kê thời gian chạy (giây) với thuật toán Hopcroft-Karp")

idx = 6
for i in range(2):
    for j in range(3):
        axs[i, j].boxplot(master[i*3+j].values[:,idx], meanline=True)
        axs[i, j].set_title("n = {0}".format(master[i*3+j].values[3,0]))
```
